```python
def x_bar_limits(data, range):
  # CENTER = data_mean
  # LCL = data_mean - A2 * range_bar
  # UCL = data_mean + A2 * range_bar
  
  CENTER = data.mean()
  LCL = CENTER - CLC[sample_size]['A2'] * range.mean()
  UCL = CENTER + CLC[sample_size]['A2'] * range.mean()
  
  return CENTER, LCL, UCL
```

```python
def range_limits(range):
  # CENTER = range_bar
  # LCL = D3 * range_bar
  # UCL = D4 * range_bar
  
  CENTER = range.mean()
  LCL = CLC[sample_size]['D3'] * range.mean()
  UCL = CLC[sample_size]['D4'] * range.mean()

  return CENTER, LCL, UCL
```

```python
def validate_out_control(data, CENTER, LCL, UCL):
  Vx = []
  Vy = []
  # Se validan los puntos fuera de limites de control
  i = 0
  control = True
  for group in data:
      if group > UCL or group < LCL:
          print('El grupo', i, 'esta fuera de los limites de control!!')
          Vx.append(i)
          Vy.append(group)
          control = False
      i += 1
  if control == True:
    print('Todos los puntos estan dentro de los limites de control')
  return Vx, Vy
```

```python
def control_plot(data, CENTER, LCL, UCL, Vx, Vy):
  fig = go.Figure()

  # Se agrega la linea de los grupos y su media
  fig.add_trace({'type': 'scatter',
                # 'x': [],
                'y': data,
                'mode': 'lines+markers',
                'name': 'Data',
                'showlegend': True,
                'hoverinfo': 'all',
                'line': {'color': 'black',
                          'width': 2},
                'marker': {'color': 'black',
                            'size': 8,
                            'symbol': 'circle'}
                })
  # Se agregan los punos fuera de control
  fig.add_trace({'type': 'scatter',
                'x': Vx,
                'y': Vy,
                'mode': 'markers',
                'name': 'Violations',
                'showlegend': True,
                'hoverinfo': 'all',
                'marker': {'color': 'rgb(255,65,54)',
                            'line': {'width': 3},
                            'opacity': 0.5,
                            'size': 12,
                            'symbol': 'circle-open'}
                })
  # Se agrega la linea central de la media de los grupos
  fig.add_trace({'type': 'scatter',
                'x': [0, len(data)],
                'y': [CENTER, CENTER],
                'mode': 'lines',
                'name': 'Center',
                'showlegend': True,
                'hoverinfo': 'all',
                'line': {'color': 'blue',
                          'width': 1}
                })
  # Se agregan las lineas de los limites de control superior e inferior
  fig.add_trace({'type': 'scatter',
                'x': [0, len(data), None, 0, len(data)],
                'y': [UCL, UCL, None, LCL, LCL],
                'mode': 'lines',
                'name': 'Limits',
                'showlegend': True,
                'hoverinfo': 'all',
                'line': {'color': 'red',
                          'width': 1,
                          'dash': 'dash'}
                })
  return fig
```

```python
def x_bar_plot(data):
  # Se obtienen los grupos, muestras y valores estadisticos del conjunto de datos
  n_groups, sample_size = data.shape
  stats = data.agg(["mean", "std", "min", "max"], axis=1)
  x_bar = stats.iloc[:,0]
  std = stats.iloc[:,1]
  range = stats.iloc[:,3] - stats.iloc[:,2]
  print("Numero de grupos:", n_groups, "Cantidad de muestras: ", sample_size)
  # Se obtienen los limites de control para la grafica de medias
  CENTER, LCL, UCL =  x_bar_limits(x_bar, range)
  # Se obtienen puntos fuera de control para la grafica de medias
  Vx, Vy = validate_out_control(x_bar, CENTER, LCL, UCL)
  # Se genera la grafica de medias
  fig = control_plot(x_bar, CENTER, LCL, UCL, Vx, Vy)
  fig.show()
```

```python
def range_plot(data):
  # Se obtienen los grupos, muestras y valores estadisticos del conjunto de datos
  n_groups, sample_size = data.shape
  stats = data.agg(["mean", "std", "min", "max"], axis=1)
  x_bar = stats.iloc[:,0]
  std = stats.iloc[:,1]
  range = stats.iloc[:,3] - stats.iloc[:,2]
  print("Numero de grupos:", n_groups, "Cantidad de muestras: ", sample_size)
  # Se obtienen los limites de control para la grafica de medias
  CENTER, LCL, UCL =  range_limits(range)
  # Se obtienen puntos fuera de control para la grafica de medias
  Vx, Vy = validate_out_control(range, CENTER, LCL, UCL)
  # Se genera la grafica de medias
  fig = control_plot(range, CENTER, LCL, UCL, Vx, Vy)
  fig.show()
```

```python
def CPK(data, LSL, USL, target=None):
  ##################################
  ########     CALCULOS     ########
  ##################################
  # Si no se especifica un objetivo este es el centro de los datos
  if target is None:
    target = min(data.min()) + ((max(data.max()) - min(data.min())) / 2.0)
  n_groups, sample_size = df.shape
  stats = df.agg(["mean", "std"], axis=1)
  mean = stats["mean"].mean()
  std = stats["std"].mean()
  total = n_groups * sample_size
  CP = (USL - LSL) / (6.0 * std)
  CPU = (USL - mean) / (3.0 * std)
  CPL = (mean - LSL) / (3.0 * std)
  CPK = min(CPU, CPL)
  overview =  pd.DataFrame(data=[total, n_groups, sample_size, std, mean, target, CP, CPU, CPL, CPK],
                           columns=['Overview'],
                           index=["Total", "Grupos", "Tamao", "DevStd", "Media", "Objetivo", "CP", "CPU", "CPL", "CPK"])
  ##################################
  ########     GRAFICA      ########
  ##################################
  data_plot = pd.DataFrame(data.stack().values, columns=['data'])
  # Se genera la grafica con la curva normal y la distribucion de los datos
  fig = ff.create_distplot([data_plot[c] for c in data_plot.columns], data_plot.columns, curve_type='normal', show_rug=False, show_hist=True)
  fig.update_traces(autobinx=True, selector={'type':'histogram'})
  fig.update_layout(bargap=0.01, autosize=False)
  
  fig.update_layout(shapes=[# Se agrega la linea de objetivo
                            dict(type='line',
                                 yref='paper',
                                 y0=0,
                                 y1=1,
                                 xref='x',
                                 x0=target,
                                 x1=target,
                                 line={'color':'red', 'width':1, 'dash':'dash'}),
                             # Se agrega la linea del limite inferior de especificacion
                            dict(type='line',
                                 yref='paper',
                                 y0=0,
                                 y1=1,
                                 xref='x',
                                 x0=LSL,
                                 x1=LSL,
                                 line={'color':'red', 'width':1, 'dash':'dot'}),
                            # Se agrega la linea del limite superior de especificacion
                            dict(type='line',
                                 yref='paper',
                                 y0=0,
                                 y1=1,
                                 xref='x',
                                 x0=USL,
                                 x1=USL,
                                 line={'color':'red', 'width':1, 'dash':'dot'})])
  # Se agregan anotaciones a las lineas con una leyenda y el valor de cada una
  fig.update_layout(annotations=[# Se agrega el valor del limite inferior de especificacion
                                 dict(text=str(LSL),
                                      align='left',
                                      yanchor='top',
                                      x=LSL,
                                      y=1.06,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9}),
                                 # Se agrega la leyenda "LSL" del limite inferior de especificacion
                                 dict(text='LSL',
                                      align='left',
                                      yanchor='top',
                                      x=LSL,
                                      y=1.12,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9}),
                                 # Se agrega el valor del objetivo
                                 dict(text=str(target),
                                      align='left',
                                      yanchor='top',
                                      x=target,
                                      y=1.06,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9}),
                                 # Se agrega la leyenda "Target" del limite inferior de especificacion
                                 dict(text='Target',
                                      align='left',
                                      yanchor='top',
                                      x=target,
                                      y=1.12,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9}),
                                 # Se agrega el valor del limite superior de especificacion
                                 dict(text=str(USL),
                                      align='left',
                                      yanchor='top',
                                      x=USL,
                                      y=1.06,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9}),
                                 # Se agrega la leyenda "USL" del limite inferior de especificacion
                                 dict(text='USL',
                                      align='left',
                                      yanchor='top',
                                      x=USL,
                                      y=1.12,
                                      showarrow=False,
                                      xref="x",
                                      yref="paper",
                                      font={'size':9})])
 
  return overview, fig
```
