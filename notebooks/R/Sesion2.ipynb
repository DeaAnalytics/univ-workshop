```r
#Leer el archivo que se llama TicketPrice_Airlines.csv y renombrarlo como "df"
df <- read.csv("https://github.com/DeaAnalytics/univ-workshop/raw/main/data/TicketPrice_Airlines.csv")
```

```r
#Ver el contenido de mi tabla llamada df
df
```

```r
#Hacer un histograma simple, con falta de formato:
hist(df$TicketPrice)
```

```r
#Ahora a mi histograma vamos a ponerle títulos entendibles y color
hist(x = df$TicketPrice, main = "Histograma de Ticket Price", xlab = "Ticket Price", 
     ylab = "Frecuencia", col="blue")
```

```r
#Puedo cambiar el color, en la línea donde dice Col, cambiamos lo que hay de lado derecho del igual (entre comillas), de la siguiente manera (todo lo demás se deja igual):
hist(x = df$TicketPrice, main = "Histograma de Ticket Price", xlab = "Ticket Price", 
     ylab = "Frecuencia", col=blues9)
```

```r
#Si solo te interesa analizar una parte del eje x, puedes acotarlo utilizando xlim de la siguiente manera:
hist(x = df$TicketPrice, main = "Histograma de Ticket Price", xlab = "Ticket Price", 
     ylab = "Frecuencia", col=blues9, xlim=c(300,500))
```

```r
#De manera muy similar si deseas acotar el eje y, se hace de la siguiente manera, solo que en lugar de xlim, ahora usamos ylim:
hist(x = df$TicketPrice, main = "Histograma de Ticket Price", xlab = "Ticket Price", 
     ylab = "Frecuencia", col=blues9, ylim=c(0,40))
```

```r
#Plus: Como hacer un gráfica de densidad sin formato:
plot(density(df$TicketPrice))
```

```r
#Como hacer un gráfica de densidad con un poco más de formato:
plot(density(df$TicketPrice), main = "Densidad de Ticket Price", xlab = "Ticket Price", 
     ylab = "Densidad", col="blue")
```

```r
#Ahora veamos como hacer un boxplot, de inicio sin formato
boxplot(df$TicketPrice)
```

```r
#Ahora hagamos más presentable nuestro boxplot, le ponemos nombre, le ponemos color y lo vamos a poner de manera horizontal:
boxplot(df, main = "Boxplot de TicketPrice", col = c("#0000CC"), notch = T, horizontal = TRUE)
summary(df)
```

```r
#Ahora, supongamos que queremos analizar estas 3 representaciones gráficas en uno solo y un summary inicial para mejor visión de los resultados:
summary(df)
par(mfrow = c(2,1))
plot(density(df$TicketPrice), main = "Densidad de Ticket Price", xlab = "Ticket Price", 
     ylab = "Densidad", col="blue")
hist(x = df$TicketPrice, main = "Histograma de Ticket Price", xlab = "Ticket Price", 
     ylab = "Frecuencia", col=blues9)
boxplot(df, main = "Box Plot de TicketPrice", col = c("#0000CC"), notch = T, horizontal = TRUE)
par(mfrow = c(1,1))
#Que puntos son considerados outlier: Q3+1.5IQR y Q#-1.5IQR (donde IQR: rango ingercuartílico = Q3-Q1)
```

```r
#Ahora vamos a leer el archivo que se llama Matriz_Correlacion.csv y renombrarlo como "df2" para hacer análisis futuros y veamos que contiene
df2 <- read.csv("https://github.com/DeaAnalytics/univ-workshop/raw/main/data/Matriz_Correlacion.csv")
df2
```

```r
#Ahora hagamos un Diagrama de Dispersión que nos permitirá si hay alguna relación con las variables de mi tabla, elegimos quien será x y y (cualesquier par de variables de mi tabla df2)
#Primero tomemos a x=Temperatura del sector1 y y=Temperatura del sector2
plot(x = df2$TempS1, y = df2$TempS2)
```

```r
#Ahora hagamos un Diagrama de Dispersión con x=Torque S1 y y=TorqueS2
plot(x = df2$TempS1, y = df2$TorqueS1)
```

```r
#Y lo mismo podemos hacer con las diferentes combinaciones en pares de las variables contenidas en la tabla:
plot(x = df2$TempS2, y = df2$TorqueS2)
```

```r
#Ahora hagamos una serie de boxplot para ver la relación de una variable respecto ala otra y le damos un poco de formato:
#Primero veo que variables quiero relacionar y lo renombro como dftable, después obtengo el boxplot correspondiente.
dftable <- data.frame(df2$TempS1, df2$TorqueS1)
boxplot(df2.TempS1 ~ df2.TorqueS1, data=dftable, main="Boxplot", xlab = "TorqueS1", ylab = "TemperaturaS1", col = c("blue", "green", "red", "pink", "yellow", "purple", "orange")) 
```

```r
#Antes de hacer una Matriz de correlación, debemo instalar los siguientes paquetes
install.packages("corrplot")
```

```r
#Y vamos a utilizar la siguiente librería:
library(corrplot)
```

```r
#Primero obtengamos una matriz de correlación simple, redondeando los coeficientes de correlación y renombrandolo como "corr"
corr <- round(cor(df2),1)
corr
```

```r
#Aquí esta otra forma d eobtener una matriz de correlación, pero interpretada de manera numérica y con figuras de círculo (entre más grande y color azul fuerte sea el círculo, la correlación también lo es)
corrplot.mixed(corr,
               lower = "number", 
               upper = "circle",
               tl.col = "black")
```

```r
#Ahora otra forma de Matriz de correlación, solo numérica y solo diagonal superior, entre más cercano este el valor a 1 o -1 mayor correlación entre las variables (directa o inversa) :

corrplot(corr, method="number", type="upper")
```

```r
#Otra forma de Matriz de correlación, entre más eliptica sea la figura, mayor correlación entra las variables:
corrplot(corr, method = "ellipse")
```
