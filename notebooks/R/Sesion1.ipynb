```r
### Continuación: Voy a instalar un paquete que me servirá más adelante para usar una librería que me permite eliminar columnas
install.packages("dplyr")
```

```r
#utilizamos la librería que nos permitirá eliminar columnas:
library("dplyr")
```

```r
#Voy a eliminar aquella(s) variable(s)/(columnas) que contienen muchos NULL y que no son de utilidad para mi estudio, en este caso "company" y la voy a renombrar como MCU2:
MCU2 <- select(MCU, -company)
#Obtengo los primeros datos de mi tabla para garantizar que la columna/variable fue eliminada, como verán ya no aparece "company"
head(MCU2)
```

```r
#Si deseo reemplazar/sustituir algún NULL, por ejemplo reemplazar los NULL por ceros en la variable "children" y renombro mi tablas para quedarme con la actualización:
MCU3 <- MCU2 %>% mutate(children = replace(children,which(is.na(children)), "0"))
MCU3
```

```r
#Hago lo mismo que en el paso anterior para la variable "" y renombro mi tabla para quedarme con la actualización:
MCU3_1 <- MCU3 %>% mutate(days_in_waiting_list = replace(days_in_waiting_list,which(is.na(days_in_waiting_list)), "0"))
MCU3_1 
```

```r
#Elimino todas las filas que contengan algún NULL
MCU4 <- na.omit(MCU3_1)
MCU4
```

```r
#Contabilizamos cuantos NULL tiene nuestra tabla MCU
sapply(MCU4, function(x) sum(is.na(x)))
```

```r
#Ahora vamos a normalizar los datos, para ajustar los valores medios en diferentes escalas respecto a una escala común, ya que hay variabes que contienen datos muy grandes respecto a otra:
#"Para eso ocuparemos los siguientes paquetes y librerías"
install.packages("kableExtra")
```

```r
install.packages("scales")
```

```r
library("kableExtra")
```

```r
library("scales")
```

```r
#Normalizamos las columnas 8 y 11 que son las que tienen datos que son muy grandes respecto al resto y lo renombramos como "Normal"
Normal <- scale(MCU4[,c(8,11)], center = T, scale = T)
kable(head(Normal), format= "markdown")
Normal
```

```r
#Ahora vamos a unir las tablas MCU4 (con datos sin NULL) y nuestra tabla (con variables normalizadas) a través de nuestra primary key y la renombramos como MCU5:
MCU5 <- cbind(MCU4, Normal)
MCU5
```

```r
#Ahora hay que eliminar las columnas 8 y 11 (que están en MCU5 pero no están normalizadas) y renombramos para actualizar:
MCU6 <- MCU5[,-c(8,11)]
MCU6
#Y listo!
```
