```r
#Leer el archivo que se encuentra en GitHub, que se llama Prueba_CPCPK_QUITARMUESTRA.csv y renombrarlo como "data"
data <- read.csv("https://github.com/DeaAnalytics/univ-workshop/raw/main/data/Prueba_CPCPK_QUITOMUESTRA.csv")
data
```

```r
#Instalamos los paquetes que utilizaremos 
install.packages("qcc")
```

```r
library(qcc)
```

```r
#Con la siguiente función, obtenemos la gráfica de medias, que como verán tenemos 2 outlier
G.Rangos <- qcc(data = data, type = "R", plot = T)
```

```r
#Obtengamos la gráfica de Medias
G.Medias <- qcc(data = data, type = "xbar", plot = T)
```

```r
#Ahora vamos a quitar los dos outlier que ví en mi gráfica de Rangos, eliminándolos de mi tabla original y ahora la llamo "Data2":
Data2 <- data[-c(6,21),]
Data2
```

```r
#Veamos como se ve la gráfica de Rangos sin outlier y le llamamos G.Rangos2
G.Rangos2 <- qcc(data = Data2, type = "R", plot = T)
```

```r
#Ahora veamos como se ve nuestra gráfica de medias sin los outlier
G.Medias2 <- qcc(data = Data2, type = "xbar", plot = T)
```

```r
#Ahora si haremos un análisis de Capacidad del Proceso: Obteniendo los valores de CP y CPK, dados los límites de especificación del producto
Capacidad <- process.capability(object = G.Medias2, spec.limits = c(1.81, 2.05))
                                Capacidad$indices
                                Capacidad$exp
                                Capacidad$obs
                                Capacidad$std.dev
```
