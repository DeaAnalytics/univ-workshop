```r
#Utilizamos la data que ya se encuentra en Github como RegresionMultiple.csv y renombramos a la tabla como "datos"
datos <- read.csv("https://github.com/DeaAnalytics/univ-workshop/raw/main/data/RegresionMultiple.csv")
datos
```

```r
#Instalamos el paquete y librería llamados "dplyr" que utilizaremos para funciones más adelante
install.packages("dplyr")
library(dplyr)
```

```r
#Definimos densidad de población = habitantes*1000/area y lo agregamos como una columna más a nuestra data con la función mutate
#renombramos todo esto nuevo como "datos" (se guarda "encima" de lo que existía antes y toma la última actualización de la tabla llamada datos)
datos <- mutate(.data = datos, Densidad_pobl = Habitantes * 1000 /Area)
datos
```

```r
#De la tabla anterior quitamos la primera columna ya que para este análisis no nos será de utilidad. Y la renombro otra vez como datos
datos <- select(datos, -c(1,))
datos
```

```r
#Analizamos los coeficientes de correlación para ver posibles colinealidades, utilizando la función "cor" y redondeando a 3 decimales con "round".
#En este caso utilizamos el método de pearson, pero hay otros que dependiendo de la naturaleza de tus datos, deberás elegir dicho método.
round(cor(x = datos, method = "pearson"), 3)
```

```r
#Instalamos el paquete y utlizamos la librería "GGally" que nos servirá para utilizar "ggpairs" más adelante
install.packages("GGally")
library(GGally)
```

```r
#ggpairs nos permite representar en un mismo escenario diagramas de dispersión, Distribuciones y coeficientes de correlación para cada par de vars.
#Usamos unas clave valor para decir que en el triángulo inferior de la tabla me coloque unos Gráficos tipo "smooth" (de dispersión)
#Usamos una clave valor para decir que en la diagnoal de la tabla coloque el Gráfico de barras a través de "barDiag" (Distribuciones)
#De manera automática en la diagonal superior se colocarán los coeficientes de correlación
ggpairs(datos, lower = list(continuous = "smooth"),
        diag = list(continuous = "barDiag"), axisLabels = "none")
```

```r
#Se genera un PRIMER MODELO que incluye TODAS las variables
#Utilizamos la función "lm" de modelo lineal, donde la primer variable dentro de los () es la variable a predecir
#Después del ~ colocamos el resto de las variables de mi tabla. Y al final un summary para que nos despliegue el ANOVA 
modelo <- lm(Esp_Vida ~ Habitantes + Ingresos + Analfabetismo + Asesinatos +
               Universitarios + Heladas + Area + Densidad_pobl, data = datos)
summary(modelo)
```

```r
#Ahora intentemos genera un SEGUNDO MODELO a través de la Selección de mejores predictores:
#utilizamos la función de step (método STEPWISE MIXTO) para poder elegir los mejores predictores
#A través del Akaike (AIC) determinaremos que modelo es mejor y por ende cuales son nuestras mejores variables predictoras
#AIC es un valor matemático que se utiliza para determinar cual es el mejor modelo.
step(object = modelo, direction = "both", trace = 1)
```

```r
#Después de ver los resultados, elegimos el mejor modelo según STEP (el que obtuvo el AIC=-28.16) 
#y obtenemos un summary para ver su ANOVA. Esto nos servirá para construir la ecuación y hacer las validaciones estadísticas
#Renombro todo como "modelo"
modelo <- (lm(formula = Esp_Vida ~ Habitantes + Asesinatos + Universitarios +
                Heladas, data = datos))
summary(modelo)
```

```r
#Instalo los paquetes y uso las librerías que utilizaré en las funciones siguientes:
install.packages("ggplot2")
install.packages("gridExtra")
library(ggplot2)
library(gridExtra)
```

```r
####COMENZAMOS LAS VALIDACIONES
```

```r
#Paso 1) Validación de condición de linealidad para todos los predictores: "Habitantes", "Asesinatos", "Universitarios" y "Heladas"
#Similar a lo que ya hemos hecho en Regresión lineal simple, con función de ggplot para ver las gráficas de Residuales
plot1 <- ggplot(data = datos, aes(Habitantes, modelo$residuals)) +
  geom_point() + geom_smooth(color = "firebrick") + geom_hline(yintercept = 0) +
  theme_bw()
plot2 <- ggplot(data = datos, aes(Asesinatos, modelo$residuals)) +
  geom_point() + geom_smooth(color = "firebrick") + geom_hline(yintercept = 0) +
  theme_bw()
plot3 <- ggplot(data = datos, aes(Universitarios, modelo$residuals)) +
  geom_point() + geom_smooth(color = "firebrick") + geom_hline(yintercept = 0) +
  theme_bw()
plot4 <- ggplot(data = datos, aes(Heladas, modelo$residuals)) +
  geom_point() + geom_smooth(color = "firebrick") + geom_hline(yintercept = 0) +
  theme_bw()
grid.arrange(plot1, plot2, plot3, plot4)
```

```r
#Con las 4 gráficas anteriores, vemos que se cumple la linealidad para todas las variables predictoras
```

```r
#Paso 2) Validación de distribución normal de los residuos con la gráfica QQPlot a través de la función "qqnorm"
#Y con "Shapiro test" para conocer su p-value
qqnorm(modelo$residuals)
qqline(modelo$residuals)
shapiro.test(modelo$residuals)
```

```r
#La prueba de Shapiro y la QQplot confirman la normalidad
```

```r
#Paso 3) Validación de homocedasticidad con ggplot (ya visto antes), para ver los residuales vs valores ajustados
#definimos quienes son los ejes con "aes",
#coloco el "geom_smooth" para colocar la linea, donde le damos color "firebrick"
#y una geom_hline para colocar sobre el cero el eje divisorio
#Finalmente uso el Breusch-Pagan test (bptest) para conocer su p-value
#uso el paquete y librería lmtest
#como se vió en la teoría, recuerden que si vemos algún patrón o mayor dispersión en algúno de los extremos de la gráfica, esto significa que no hay homocedasticidad
#En esta gráfica recordemos que se deben ver valores distribuidos de manera aleatoria alrededor del cero.
ggplot(data = datos, aes(modelo$fitted.values, modelo$residuals)) +
  geom_point() +
  geom_smooth(color = "firebrick", se = FALSE) +
  geom_hline(yintercept = 0) +
  theme_bw()
install.packages("lmtest")
library(lmtest)
bptest(modelo)
```

```r
#En la gráfica anterior, no hay indicios de homocedasticidad
```

```r
#Paso 5)Detección de posibles datos atípicos o influyentes con gráfico de Distribución de los residuos estandarizados
#para esto necesitamos el paquete (ya instalado) y la librería "dplyr"
#con ggplot hacemos el gráfico de Dispersión de los residuales estandarizados
#colocamos una geom_hline para graficar una línea recta en y=3 y le pongo color verde y linetype= dashed que es punteada
#con geom_point, se identifican en rojo observaciones con residuos estandarizados absolutos > 3 y si no, en color azul
#finalmente coloco títulos y nombres a los ejes.
library(dplyr)
datos$studentized_residual <- rstudent(modelo)
ggplot(data = datos, aes(x = predict(modelo), y = abs(studentized_residual))) +
  geom_hline(yintercept = 3, color = "green", linetype = "dashed") +
  geom_point(aes(color = ifelse(abs(studentized_residual) > 3, 'red', 'blue'))) +
  scale_color_identity() +
  labs(title = "Distribución de los residuos studentized",
       x = "predicción") + 
  theme_bw() + theme(plot.title = element_text(hjust = 0.5))
```
